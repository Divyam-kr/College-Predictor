```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("CET_Database_Final2019.csv")
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
fixed_columns = ['CETCode', 'College', 'Location', 'Branch']
category_columns = [col for col in df.columns if col not in fixed_columns]

# Step 2: Melt to long format
df_melted = df.melt(
    id_vars=['College', 'Location', 'Branch'],
    value_vars=category_columns,
    var_name='Category',
    value_name='Cutoff_Rank'
)

# Step 3: Convert to numeric & filter valid entries
df_melted['Cutoff_Rank'] = pd.to_numeric(df_melted['Cutoff_Rank'], errors='coerce')
df_cleaned = df_melted[df_melted['Cutoff_Rank'] > 0].reset_index(drop=True)

# Step 4: Final features and label
final_df = df_cleaned[['Category', 'Cutoff_Rank', 'Branch', 'Location', 'College']]  # College is the target
final_df
```

```python
final_df
```

```python
final_df["Location"].value_counts()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import pickle


college_counts = final_df['College'].value_counts()
valid_colleges = college_counts[college_counts >= 2].index
final_df = final_df[final_df['College'].isin(valid_colleges)]

X = final_df[['Category', 'Cutoff_Rank', 'Branch', 'Location']]
y = final_df['College']

category_encoder = OneHotEncoder(sparse=False, drop='first')
branch_encoder = OneHotEncoder(sparse=False, drop='first')
location_encoder = OneHotEncoder(sparse=False, drop='first')
rank_scaler = StandardScaler()

X_category = category_encoder.fit_transform(X[['Category']])
X_branch = branch_encoder.fit_transform(X[['Branch']])
X_location = location_encoder.fit_transform(X[['Location']])
X_rank = rank_scaler.fit_transform(X[['Cutoff_Rank']])

X_transformed = pd.concat([
    pd.DataFrame(X_category, index=X.index),
    pd.DataFrame(X_branch, index=X.index),
    pd.DataFrame(X_location, index=X.index),
    pd.DataFrame(X_rank, index=X.index)
], axis=1)

X_train, X_test, y_train, y_test = train_test_split(
    X_transformed, y, test_size=0.2, random_state=42
)

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

with open('kcet_trained_model.pkl', 'wb') as f:
    pickle.dump(clf, f)

print("✅ Trained model saved as 'kcet_trained_model.pkl'")



with open('kcet_encoders_scaler.pkl', 'wb') as f:
    pickle.dump({
        'category': category_encoder,
        'branch': branch_encoder,
        'location': location_encoder,
        'scaler': rank_scaler
    }, f)
print("✅ Encoder model saved as 'kcet_encoder_model.pkl'")
```
